## 1.Sphere

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 100)
Theta, Phi = np.meshgrid(theta, phi)

r = 1  # شعاع کره
X = r * np.sin(Phi) * np.cos(Theta)
Y = r * np.sin(Phi) * np.sin(Theta)
Z = r * np.cos(Phi)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Sphere")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

## 2.Ellipsoid

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 100)
Theta, Phi = np.meshgrid(theta, phi)

a, b, c = 1, 1.5, 2  # نیم‌محورها
X = a * np.sin(Phi) * np.cos(Theta)
Y = b * np.sin(Phi) * np.sin(Theta)
Z = c * np.cos(Phi)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Ellipsoid")
plt.show()
```

## 3.Elliptic Hyperboloid of One Sheet

```python
import numpy as np
import matplotlib.pyplot as plt
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(-2, 2, 100)
Theta, Z = np.meshgrid(theta, z)
X = np.cosh(Z) * np.cos(Theta)
Y = np.cosh(Z) * np.sin(Theta)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Hyperboloid of One Sheet')
plt.show()
```

## 4.Elliptic Hyperboloid of Two Sheets

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)
z = np.linspace(-2, -0.1, 50)
z2 = np.linspace(0.1, 2, 50)
Theta, Z1 = np.meshgrid(theta, z)
Theta, Z2 = np.meshgrid(theta, z2)

a, b, c = 1, 1, 1

X1 = np.sqrt((Z1**2 / c**2) - 1) * a * np.cos(Theta)
Y1 = np.sqrt((Z1**2 / c**2) - 1) * b * np.sin(Theta)

X2 = np.sqrt((Z2**2 / c**2) - 1) * a * np.cos(Theta)
Y2 = np.sqrt((Z2**2 / c**2) - 1) * b * np.sin(Theta)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, Y1, Z1, cmap='coolwarm')
ax.plot_surface(X2, Y2, Z2, cmap='coolwarm')
ax.set_title("Hyperboloid of Two Sheets")
plt.show()
```

## 5.Elliptic Cone

```python
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(-2, 2, 100)
Theta, Z = np.meshgrid(theta, z)

a, b, c = 1, 1, 1
R = np.abs(Z)
X = a * R * np.cos(Theta)
Y = b * R * np.sin(Theta)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='spring')
ax.set_title("Elliptic Cone")
plt.show()
```

## 6.Elliptic Paraboloid

```python
u = np.linspace(0, 2 * np.pi, 100)
r = np.linspace(0, 2, 100)
U, R = np.meshgrid(u, r)

a, b = 1, 1
X = R * np.cos(U)
Y = R * np.sin(U)
Z = (X**2 / a**2) + (Y**2 / b**2)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='inferno')
ax.set_title("Elliptic Paraboloid")
plt.show()
```

## 7.Hyperbolic Paraboloid

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

a, b = 1, 1
Z = (X**2 / a**2) - (Y**2 / b**2)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("Hyperbolic Paraboloid")
plt.show()
```

## 8.Cylinder

```python
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(-2, 2, 100)
Theta, Z = np.meshgrid(theta, z)

a, b = 1, 2
X = a * np.cos(Theta)
Y = b * np.sin(Theta)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='autumn')
ax.set_title("Elliptic Cylinder")
plt.show()
```
